### Import Libs

```python
import cv2
from deepface import DeepFace
import matplotlib.pyplot as plt
```

### Testing

```python
def get_emotion(img):
    try:
        imageFace = cv2.imread(img)
        predictions = DeepFace.analyze(imageFace, actions = ['emotion'])
        return predictions
    except ValueError:
        return "No face detected"


if __name__ == '__main__':
    images = ['testing_images/img1.jpg', 
              'testing_images/img2.jpg',
              'testing_images/img3.jpg',
              'testing_images/img4.jpg',
              'testing_images/img5.jpg',
              'testing_images/img6.jpg',]
    
    for img in images:
        emotion = get_emotion(img)
        if emotion != "No face detected":
            print(*emotion[0]['dominant_emotion'].split('_'), sep=' ')
            plt.imshow(cv2.cvtColor(cv2.imread(img), cv2.COLOR_BGR2RGB))
            plt.show()
        else:
            print(emotion)
```
