### Import Libs

```python
from fer import FER
import cv2
import matplotlib.pyplot as plt
```

### Testing

```python
def get_emotion(img):
    emo_detector = FER(mtcnn=True)
    dominant_emmotion, emotion_score = emo_detector.top_emotion(img)
    return dominant_emmotion, emotion_score


def show_emotion(img):
    dominant_emmotion, emotion_score = get_emotion(img)
    plt.imshow(img[:,:,::-1])
    plt.title(f"Emotion: {dominant_emmotion} with score: {emotion_score}")
    plt.show()


if __name__ == "__main__":
    images = ['testing_images/img1.jpg', 
              'testing_images/img2.jpg',
              'testing_images/img3.jpg',
              'testing_images/img4.jpg',
              'testing_images/img5.jpg',
              'testing_images/img6.jpg',]
    
    for img in images:
        show_emotion(cv2.imread(img))
```
